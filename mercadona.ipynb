```python
import requests

dic = {}
url = 'https://tienda.mercadona.es/api/categories/?lang=es&wh=vlc1'
respuesta = requests.get(url)
data = respuesta.json()

categorias = data['results']

for elem in categorias:
    categoria_name = elem['name']
    sub = elem['categories']

    if categoria_name not in dic:
        dic[categoria_name] = {}

    for elem2 in sub:
        idSub = elem2['id']
        url = f'https://tienda.mercadona.es/api/categories/{idSub}/?lang=es&wh=vlc1'
        subcategorias = requests.get(url)
        r = subcategorias.json()['categories']

        for i in r:
            for j in i['products']:
                producto_id = j['id']
                name = j['display_name']
                precio = j['price_instructions']['reference_price']

                dic[categoria_name][producto_id] = {
                    'nombre': name,
                    'precio': precio
                }

print(dic)
```

```python
import csv

with open("mercadona_precios.csv", mode="w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)

    writer.writerow(["Categoria", "ID Producto", "Producto", "Precio"])

    for categoria, productos in dic.items():
        for producto_id, info in productos.items():
            writer.writerow([
                categoria,         
                producto_id,       
                info['nombre'],    
                info['precio']     
            ])
```
