```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
url = "https://www.compraonline.alcampo.es/categories/bebidas/agua-soda-y-gaseosas/aguas-sin-gas/OC1101041?source=navigation"
driver.get(url)
driver.implicitly_wait(3)
```

```python
coockie_button = driver.find_element(By.ID, "onetrust-accept-btn-handler")
coockie_button.click()
```

```python
box = driver.find_element(By.CLASS_NAME, "box-close")
box.click()
```

```python
productos = driver.find_elements(By.CLASS_NAME, "product-card-container")

diccionario_productos = {}

for producto in productos:
    try:
        nombre = producto.find_element(By.CSS_SELECTOR, "h3[data-test='fop-title']").text
        precio = producto.find_element(By.CSS_SELECTOR, "span[data-test='fop-price']").text
        diccionario_productos[nombre] = precio
    except:
        pass
```

```python
print (diccionario_productos)
```

```python
import csv

with open("alcampo_productos.csv", mode="w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter=";")
    writer.writerow(["Categoria", "Producto", "Precio"])
    for categoria, productos in diccionario_final.items():
        for producto, precio in productos.items():
            writer.writerow([categoria, producto, precio])
```
