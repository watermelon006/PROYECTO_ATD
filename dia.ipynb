```python
import requests
url_api = "https://www.dia.es/api/v1/plp-insight/initial_analytics/l1/L"
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json",
    "Referer": "https://www.dia.es/",
    "Origin": "https://www.dia.es"}
categorias = {}
for i in range(101,127):
    if i != 124:
        url = url_api + str(i)
        respuesta = requests.get(url, headers=headers)
        respuesta.raise_for_status()
        data = respuesta.json()
        for id, seccion in data["page_product_analytics"].items():
            for producto in seccion["products"].values():
                if producto["item_category"] not in categorias:
                    categorias[producto["item_category"]] = {}
                else:
                    categorias[producto["item_category"]][producto["item_name"]] = producto["price"]

print(categorias)
```

```python
import csv

with open("dia_productos.csv", mode="w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter=";")
    writer.writerow(["Categora", " Producto", " Precio"])
    for categoria, productos in categorias.items():
        for producto, precio in productos.items():
            writer.writerow([str(categoria), str(producto), str(precio)])
```
